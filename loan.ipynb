#### Handling Missing Values  


```python
df.isnull().sum()
```

```python
# Categorical data
df['Gender'].fillna(df['Gender'].mode()[0],inplace=True)
df['Married'].fillna(df['Married'].mode()[0],inplace=True)
df['Dependents'].fillna(df['Dependents'].mode()[0],inplace=True)
df['Self_Employed'].fillna(df['Self_Employed'].mode()[0],inplace=True)
df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0],inplace=True)
df['Credit_History'].fillna(df['Credit_History'].mode()[0],inplace=True)

# Continous data
df['LoanAmount'].fillna(df['LoanAmount'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

### Converting Categories to Numerical Values

```python
df.head()
```

```python
# df['Gender'].value_counts()
# df['Married'].value_counts()
#df['Education'].value_counts()
#df['Self_Employed'].value_counts()
#df['Property_Area'].value_counts()
df['Loan_Status'].value_counts()
```

```python
df['Gender'] = df['Gender'].map({'Male':0,'Female':1})
df['Married'] = df['Married'].map({'Yes':0,'No':1})
df['Education'] = df['Education'].map({'Graduate':0,'Not Graduate':1})
df['Self_Employed'] = df['Self_Employed'].map({'No':0,'Yes':1})
df['Property_Area'] = df['Property_Area'].map({'Semiurban':0,'Urban':1,'Rural':2})
df['Loan_Status'] = df['Loan_Status'].map({'Y':1,'N':0})
```

```python
df.head()
```

### Bringing all values to the range 0 and 1

####      Xnorm=X-Xmin/Xmax-Xmin

```python
df['Dependents'] = df['Dependents'].replace('3+',3).astype(int)
df['Dependents'] = (df['Dependents']-df['Dependents'].min())/(df['Dependents'].max()-df['Dependents'].min())
```

```python
df.dtypes
```

```python
for i in df.columns[1:]:
    df[i] = (df[i] - df[i].min())/(df[i].max()-df[i].min())
```

```python
df.head()
```

```python
df.to_csv('loan_prediction_data.csv',index=False)
```

##### 1.Loading the dataset


```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
%matplotlib inline
```

```python
data = pd.read_csv('loan_prediction_data.csv')
data
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
data = data.drop('Loan_ID',axis=1)
data
```

```python
# independent variables
X = data.drop('Loan_Status',axis=1)

# Dependent Variables
y = data['Loan_Status']
```

```python
X.shape, y.shape
```

##### 2.Creating training and Validation Set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=data['Loan_Status'], test_size=0.2, random_state=10)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

##### 3.Defining the Architecture of the Model

```python
import keras
import tensorflow as tf
```

```python
from keras.models import Sequential
from keras.layers import InputLayer,Dense
```

```python
X_train.shape
X_test.shape[1]

# Defining the input neurons
input_neurons = X_train.shape[1]
```

```python
# Since loan prediction is a binary classification problem, we will haaaaave single neuron in the output layer


# Defining the number of output neurons
output_neurons = 1
```

```python
# defining hidden layers and neurons in each layer

number_of_hidden_layers = 2
neuron_hidden_layer_1 = 10
neuron_hidden_layer_2 = 5
```

```python
# activation function of different layers

# lets pick relu as an activation function for hidden layers
# since it is a binary classification problem, we can use sigmoid activation function in the final layer

model = Sequential()
model.add(InputLayer(input_shape=(input_neurons,)))
model.add(Dense(units=neuron_hidden_layer_1,activation='relu'))
model.add(Dense(units=neuron_hidden_layer_2,activation='relu'))
model.add(Dense(units=output_neurons,activation='sigmoid'))
```

```python
model.summary()
```

```python
# Number of parameters between input and first hidden layer
input_neurons*neuron_hidden_layer_1
```

```python
# Number of parameters between input and first hidden layer

# adding the bias for each neuron of first hidden layer

input_neurons*neuron_hidden_layer_1 + 10
```

```python
# Number of parameters between first and second hidden layer

neuron_hidden_layer_1*neuron_hidden_layer_2 + 5
```

```python
# Number of parameters between second hidden and output layer

neuron_hidden_layer_2*output_neurons + 1
```

##### 4.Compling the model(defining loss function, optimer)

```python
# compiling the model

# loss as binary_crossentripy, since we have binary classification problem
# defining the optimizer as adam
# Evaluation metric as accuracy

model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

##### 5.Training the model

```python
model_history = model.fit(X_train,y_train,validation_data=(X_test,y_test), epochs=50)
```

##### 6.Evaluating model performance on validation set

```python
# getting prediction fro the validation set

prediction = np.where(model.predict(X_test) < 0.5, 0, 1)
```

```python
# calculating the accuracy on validation set

accuracy_score(y_test, prediction)
```
